```python
# plot it in red
gdf[gdf['$900 to $999'] > 90].plot(figsize=(12,10),color="blue")
```

```python
#Okay, I will add a basemap now
```

```python
# reproject to Web Mercator
gdf_web_mercator = gdf.to_crs(epsg=3857)
```

```python
# use subplots that make it easier to create multiple layered maps
fig, ax = plt.subplots(figsize=(15, 15))

# add the layer with ax=ax in the argument 
gdf_web_mercator[gdf_web_mercator['No cash rent'] > 75].plot(
                                                               ax=ax, 
                                                               color="blue",
                                                               alpha=0.5
)

# turn the axis off
ax.axis('off')

# set a title
ax.set_title('Block Groups where 75% of its Households Do Not Pay Rent',fontsize=24,pad=20)

# add a basemap
ctx.add_basemap(ax)
```

```python
# use subplots that make it easier to create multiple layered maps
fig, ax = plt.subplots(figsize=(15, 15))

# add the layer with ax=ax in the argument 
gdf_web_mercator[gdf_web_mercator['$450 to $499'] > 90].plot(
                                                               ax=ax, 
                                                               color="red",
                                                               alpha=0.5
)

# turn the axis off
ax.axis('off')

# set a title
ax.set_title('Block Groups with more than 90% Houses that Pay $450 to $499 in Gross Rent',fontsize=24,pad=20)

# add a basemap
ctx.add_basemap(ax)
```
